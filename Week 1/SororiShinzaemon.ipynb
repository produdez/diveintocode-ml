# Introduction
## Sorori Shinzaemon problem
When Hideyoshi offered him a reward, Shinzaemon was asked what he would like to receive, and he asked for one grain of rice today, two grains the next day, four grains the day after that, and so on, doubling the amount of rice each day for 100 days. Thinking that a grain of rice would be no big deal, Hideyoshi readily agreed, but realized halfway through that doubling the amount of rice each day would result in a huge amount of rice after 100 days, so he had the reward changed to something else.

```python
#Initial imports and constants, datas
import matplotlib.pyplot as plt

#predefined functions
def plot_scatter(data_list, title , x_label, y_label):
    # %matplotlib inline
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.plot(data_list)
    plt.show()
```

# Problem 1
## Rice on 100th day

```python
def problem1(): #Calc rice on 100th day and show graph
    days = 30
    list_additional_rice = [1]
    list_cumulative_rice = [1]
    for i in range(days -1 ):
        list_additional_rice.append(list_additional_rice[-1]*2)
        list_cumulative_rice.append(list_additional_rice[-1] + list_cumulative_rice[-1])
    print('Additional rice on day 100th: ', list_additional_rice[days-1])
    print('Total rice on day 100th: ', list_cumulative_rice[days-1])
    
    plot_scatter(list_additional_rice,'Additional Rice Per Day', 'Day', 'Rice Count')
    plot_scatter(list_cumulative_rice,'Rice Total Per Day', 'Day', 'Rice Count')

problem1()
```

**Note: ** Notice that the cummulative rice is around 10 times bigger than the additional rice at day 100

# Problem 2
##  Rice Count outside 100 days

```python
def compute_sorori_shinzaemon(days=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    days: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_additional_rice : list
        List of the number of rice you get paied some day (n) (the length of the list is equal to day)
    list_cumulative_rice : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    list_additional_rice = [1]
    list_cumulative_rice = [1]
    for i in range(days -1 ):
        list_additional_rice.append(list_additional_rice[-1]*2)
        list_cumulative_rice.append(list_additional_rice[-1] + list_cumulative_rice[-1])
    return list_additional_rice,list_cumulative_rice
def problem2():
    days = 10
    list_additional_rice, list_cumulative_rice = compute_sorori_shinzaemon(days = days)
    print('Additional rice on day {}th: '.format(days), list_additional_rice[-1])
    print('Total rice on day {}th: '.format(days), list_cumulative_rice[-1])
    
    plot_scatter(list_additional_rice,'Additional Rice Per Day', 'Day', 'Rice Count')
    plot_scatter(list_cumulative_rice,'Rice Total Per Day', 'Day', 'Rice Count')
problem2()
```

# Problem 3
## Days to live with some rice

```python
RICE_PER_DAY = 30000 #grain
def days_to_live(people_count, rice_count):
    return rice_count//(people_count * RICE_PER_DAY)
def problem3():
    class_population = 30
    sorori_days = 30
    rice_count = compute_sorori_shinzaemon(sorori_days)[1][-1] #100 days of sorori shinzaemon
    print('For {} days of sorori shinzaemon worth of rice ({} grains), our class of {} people will last: {} days'.format(sorori_days,rice_count,class_population, days_to_live(class_population,rice_count)))

problem3()
```
