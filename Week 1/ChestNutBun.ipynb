# Problem
## The day when chestnut buns cover the solar system


```python
def total_time(repetition, interval):
    return repetition * interval
def chessnut_reproduction_count(total_volume, initial_volume):
    return ceil(log(total_volume/initial_volume + 1,2))
def cumulative_exponent_data(n, init):
    n_day = [init]
    data = [init]
    for x in range(n -1):
        n_day.append(n_day[-1] * 2)
        data.append(n_day[-1] + data[-1])
    return data
def plot_scatter(data, title):
    plt.title(title)
    plt.xlabel('reproduction count')
    plt.ylabel('volume[m3]')
    plt.plot(data)
    plt.show()
def test():
    print(chessnut_reproduction_count(20,2))

def generic_chessnut_problem(sample_volume, room_volume, reproduce_interval):
    reps = chessnut_reproduction_count(room_volume, sample_volume)
    time_taken = total_time(reps, reproduce_interval)
    return reps, time_taken, cumulative_exponent_data(reps,sample_volume)
def problem():
    #NOTE: chessnut filling solar
    CHESSNUT_VOL = 3e-7
    SOLAR_VOL = 9.94994941972e+16
    REPRODUCTION_INTERVAL = 5 #mins

    reps1, time1, data1 = generic_chessnut_problem(CHESSNUT_VOL, SOLAR_VOL,REPRODUCTION_INTERVAL)
    print('A chessnut bun of {}m3 takes {}mins ({} replications) to fill the solar system of {}m3'.format(CHESSNUT_VOL,time1,reps1,SOLAR_VOL))
    plot_scatter(data1,'Chessnut fill solar')

    #NOTE: ball filling tokyo dome
    BALL_VOL = 5e-7
    TOKYO_DOME_VOL = 1.24e+6
    REPRODUCTION_INTERVAL = 5 #mins

    reps2, time2, data2 = generic_chessnut_problem(BALL_VOL, TOKYO_DOME_VOL,REPRODUCTION_INTERVAL)
    print('A ball bun of {}m3 takes {}mins ({} replications) to fill the tokyo dome of {}m3'.format(BALL_VOL,time1,reps2,TOKYO_DOME_VOL))
    plot_scatter(data2,'Ball fill Tokyo Dome')



problem()
```

### Explaination for the formula:

$f(x) = \alpha * 2^{x-1}$, $ (x>=1)$ with $x$ is the current day (repetition)

Cumulative Sum: $F(n) = \alpha \sum_{x=1}^{n}f(x) = \alpha \sum_{i=0}^{n-1}{i^2} = \alpha(2^n -1)$, $i>=0$ with $n$ as the current day
